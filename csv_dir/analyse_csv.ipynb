```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
csv_files_list = glob.glob("*.csv")
```

```python
len(csv_files_list)
```

```python
csv_files_list = csv_files_list[18:]
```

```python
csv_files_dict = {}
for file in csv_files_list:
    csv_files_dict[file] = pd.read_csv(file,header=None,names=['train_acc','valid_acc'])
```

```python
plt.rcParams['figure.figsize'] = [20, 10]
for key,val in csv_files_dict.items():
    plt.plot(val.valid_acc,label=key)
    plt.legend(loc="upper left")
    plt.title('Validation Accuracy')
plt.show()
```
